# Advanced Data Analysis Using Pandas


```python
#Joining The dataframes 

# #When The data is present in various data frames we can combine that data into one data frame that can provide overall view of 
# the data frames beign combined

#This process of combining the various data frames is known as 'joining' or merging the data 
```

```python
# Let us create two dataframes 
import pandas as pd
data1={
    'Emp_name':["Karna","Karthik","Arjun"],
    'Age':[30,24,22]
}

Emp_age=pd.DataFrame(data1)
```

```python
Emp_age #Tabel 1
```

```python
data2={
    'Emp_name':["Karna","Karthik","Sunny"],
    'Emp_salary':[55,60,54]
}

Emp_salary=pd.DataFrame(data2)
```

```python
Emp_salary
```

```python
#Know We will use the different types of data merging concepts here 

#Sql joins concept 
#1) Inner Join
#2) Outer join 
#3)left Join
#4)Right Join

# Inner join can be defined as the kind of the join where only common element of the two dataframes can be joined but not the 
# other ones

#outer join so the outer join can be described as the kind of the join which joins both the table values and the missing values 
# of the particular data item can be considered as the NaN Value of the Combined Data

# #Left join : Here we have made a two tables which are Emp_age, Emp_salary Here Emp_age is left side and Emp_salary but the left
# join states that The Data Which is on the left side will be considered in the tabel 

#Right join :Here The right join is completly opposite to the left join the right side of the data will be joinnned 
```

```python
#Let us see the joins with the programming example 
pd.merge(Emp_age,Emp_salary,on='Emp_name',how='inner') #This is for inner join where the common column is based on Emp_name
```

```python
pd.merge(Emp_age,Emp_salary,on='Emp_name',how='outer') #This is for outer join where common column is based on Emp_name
```

```python
pd.merge(Emp_age,Emp_salary,on='Emp_name',how='left') #This is for left join
```

```python
pd.merge(Emp_age,Emp_salary,on='Emp_name',how='right') #this is for right join
```

```python
#what happens if there is no common column then how to join the dataframes ?
#In the above examples we have considered Emp_name as the common column 
#But if there is no common column present then we use the column_data which contains the similiarites 
Emp_salary.rename(columns={'Sports_person':'Ordinary_person'},inplace=True) 
#Here we have renamed the Emp_name to Sports_person 
#sports_person to Ordinary Person 

#so here the main motive is to change the column_names and combine the dataframes with similiar column_data 
```

```python
Emp_salary
```

```python
pd.merge(Emp_age,Emp_salary,left_on='Emp_name',right_on='Ordinary_person',how='left')
#left join based on the similar column data
```

```python
pd.merge(Emp_age,Emp_salary,left_on='Emp_name',right_on='Ordinary_person',how='right')
#right join based on the similar column data
```

```python
pd.merge(Emp_age,Emp_salary,left_on='Emp_name',right_on='Ordinary_person',how='inner')
#inner join based on the similar column data
```

```python
pd.merge(Emp_age,Emp_salary,left_on='Emp_name',right_on='Ordinary_person',how='outer')
#outer join based on the similar column data
```
