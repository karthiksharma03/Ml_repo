```python
#Creating an DataFrame using python tuples 
data=[('Karthik',20,'B.tech',3,'Data-Science'),('arjun',13,'School',8,'B-Section'),('sohel',19,'B.tech',3,'Data-science'),
     ('Saketh',20,'B.tech',3,'Data-science'),('Venky',18,'B.tech',3,'Data-science'),('Datta',20,'B.tech',3,'AI-ML')]
```

```python
combine=pd.DataFrame(data,columns=["Name","Age","Course","Year/class","Field/Type"])
```

```python
combine
```

```python
#Viewing a DataFrame or acccessing A elemnts in the data frame using iloc(),loc() functions

#loc() is known as locate function which is also used for locating the Rows and columns based on the columns labels
combine.loc[:,["Name","Age"]]

#Here we can specify the number of rows and columns using the loc() function but rember that we can access the column name 
#only through column labels 

combine.loc[0:3,["Name","Age","Course"]]

#So the above two given are the example for the loc() function 
```

```python
#iloc() is an another function which is used to view the data based on integer index

combine.iloc[0:2,0:5]
combine.iloc[1,0:5]#Accessing only one row and all columns 
combine.iloc[3:0:-1,:] #This is used to view the data from Third row to the first row in reverse order including all columns
```

```python
#If we want to retrive the last row then 
combine.iloc[-1]
```

```python
#To retrive the last column
combine.iloc[:,-1]
```

```python
combine.iloc[1,4] #To obtain an single element 
```

```python
#Operations on the dataframe 
combine.shape #Which gives the (rows,columns) of a particular dataset
```

```python
#if we want the shape of the row and column separate then 
r,c=combine.shape
print("The shape of the rows is :",r)
print("The shape of the columns is :",c)
```

```python
#Retreveing rows of the dataset 

combine[0:2] #first two rows of the dataset 
```

```python
combine["Name"] #We can retrive the column in the single column or multicolumns 
```

```python
#Finding the maximum and ,min values of the particular colummn 
print("The max value of the age column is :",combine["Age"].max())
print("The min value of the age column is :",combine["Age"].min())
```

```python
#We can also use the describe function to get all the statistic information about the dataset 
combine.describe() #But This function is only applicable for numerical columns
```

```python
#Performing the small Querys on the dataset 
#We can call the rows based on the quries also 
combine[combine.Age>18] #here we have just used a query where the age is greater than 18 we called those rows
```

```python
combine[combine.Age==combine.Age.max()]
#here we can call the rows of the age which has maximum age value 
```

```python
#We can also call the specific columns with a given query
combine[['Name','Age','Course']][combine.Age==combine.Age.max()]

#calling The particular rows with maximum age and calling only Name,Age,Course colums 
```

```python
#Sorting The data 
combine1=combine.sort_values(['Year/class'],ascending=True)
combine1

#we can also sort out the multiple columns with the help of single sort_values funcstion 
```
